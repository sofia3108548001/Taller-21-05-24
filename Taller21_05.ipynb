```python
#1 Se importan las bibliotecas necesarias para ejecutal el codigo de forma eficiente.
#Las bibliotecas que se deben instalar son numpy y matplotlib.pyplot.
# mpl_toolkits.mplot3d Esta clase se utiliza para crear gr√°ficos tridimensionales en Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 2  Define las ecuaciones del sistema 3x3 en forma de matrices A y B
A = np.array([[3, 9, -10], [1, -6, 4], [10, -2, 8]])
B = np.array([24, -4, 20])

# 3 Utiliza la funci√≥n np.linalg.solve() de NumPy para resolver el sistema de ecuaciones lineales.
# Lo que hace esta funcion es calcular la solucion exacta de la ecuacion lineal
solution = np.linalg.solve(A, B)

#4  Imprimir la soluci√≥n del sistema de ecuaciones
print("La soluci√≥n del sistema de ecuaciones es:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])

#5 Crea una figura y un eje tridimensional utilizando matplotlib.pyplot.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') #Agregar un subplot tridimensional a una figura.
#6 Genera una malla de puntos x e y utilizando np.linspace().
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
#7 Se crea una cuadricula de puntos en dos dimensiones.
#Toma estos arreglos y devuelve dos matrices, X e Y, que contienen todas las combinaciones posibles de los valores en x e y.
# Estas matrices se utilizan para evaluar las ecuaciones de los planos Z1, Z2 y Z3.

X, Y = np.meshgrid(x, y) #cuadricula bidimensional  vectores de coordenadas
Z1 = (-A[0,0]*X - A[0,1]*Y + B[0]) / A[0,2]
Z2 = (-A[1,0]*X - A[1,1]*Y + B[1]) / A[1,2]
Z3 = (-A[2,0]*X - A[2,1]*Y + B[2]) / A[2,2]
#Grafica las superficies definidas por los puntos (x, y, z)
ax.plot_surface(X, Y, Z1, alpha=0.5)
ax.plot_surface(X, Y, Z2, alpha=0.5)
ax.plot_surface(X, Y, Z3, alpha=0.5)

# Graficar el punto de intersecci√≥n
ax.scatter(solution[0], solution[1], solution[2], color='red', label='Punto de intersecci√≥n')

#Establece las etiquetas de los ejes x, y, y z
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.legend()
#Muestra el gr√°fico utilizando plt.show().
plt.show()
```
